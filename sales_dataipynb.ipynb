```python
import datetime

reference_date = sales_data['ORDERDATE'].max() + datetime.timedelta(days=1)
print(reference_date)
```

```python
rfm = sales_data.groupby('CUSTOMERNAME').agg(
    Recency = ('ORDERDATE', lambda x: (reference_date - x.max()).days),
    Frequency = ('ORDERNUMBER', 'nunique'),
    Monetary = ('SALES', 'sum')
).reset_index()

print(rfm.head(10))
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], q=4, labels=[4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), q=4, labels=[1,2,3,4])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], q=4, labels=[1,2,3,4])

rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

print(rfm.head(10))
```

```python
def segment(score):
    if score == '555':
        return 'Champions'
    elif score[0] in ['4','5'] and score[1] in ['4','5']:
        return 'Loyal Customers'
    elif score[0] in ['4','5'] and score[1] in ['1','2']:
        return 'Recent Customers'
    elif score[0] in ['1','2'] and score[1] in ['4','5']:
        return 'At Risk'
    elif score == '111':
        return 'Lost'
    else:
        return 'Potential'

rfm['Segment'] = rfm['RFM_Score'].apply(segment)
print(rfm['Segment'].value_counts())
```
